```python
import yolov1
import utils
import cv2
import numpy as np 
from time import time 
import os 

yolov1 = yolov1.YOLOV1(3)
model = yolov1.build_model()
model.load_weights('weights\\dog_cat_duck_2.h5')
```

```python
import data_loader
test_img_dir = 'D:\\projects_test\\yolov1\\datasets\\dog_cat_duck\\test' 
# test_img_dir = 'D:\\projects\\_datasets\\covid\\test_5000\\Normal'
x_test, y_test = data_loader.load_dataset(test_img_dir, (448,448), yolov1.S, yolov1.C)
```

```python
total = len(x_test)

labels = ['dog', 'cat', 'duck'] 

for i in range(total):
    start = time()
    img = (x_test[i]).copy()
    predict = model.predict(np.expand_dims(img, axis=0))
    bbox_info_list = utils.get_bbox_info_list_from_predict_single_image(predict_result=predict[0], labels=labels)
    nms = utils.do_nms(bbox_info_list=bbox_info_list, iou_th=0.5)
    out = utils.draw_rectangle(img, nms)
    print(f'{i}.dealy: {(time() - start) * 1000:.3f} ms')

    cv2.namedWindow('img', 0)
    cv2.imshow('img', out)
    k = cv2.waitKey(0)
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```
